Give Web scraping code snippet in Python that I can run from Colab Notebook to fetch data from an URL and load into a Pandas data-frame. Please ensure that the URL given is valid and it allows web-scraping.

```python
# Step 1: Install necessary libraries
!pip install requests beautifulsoup4 pandas

# Step 2: Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 3: Fetch and parse the webpage
url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 4: Extract table data
tables = soup.find_all('table', {'class': 'wikitable'})
table = tables[0]  # Assuming the first table is the one we need

rows = table.find_all('tr')
data = []

for row in rows:
    cols = row.find_all(['td', 'th'])
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Step 5: Load data into a Pandas DataFrame
df = pd.DataFrame(data)
df.columns = df.iloc[0]  # Set the first row as the header
df = df[1:]  # Remove the first row from the data
df.reset_index(drop=True, inplace=True)

# Step 6: Display the DataFrame
df.head()
```
